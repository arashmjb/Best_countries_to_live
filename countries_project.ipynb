```python
# pie chart of the best countries to live based on the regional 
countries[(countries["Healthy life expectancy"] > 72)  & (countries["Social support"]> 0.90) &  (countries["Freedom to make life choices"]> 0.90)]["Regional indicator"].value_counts().plot(kind="pie")
```
